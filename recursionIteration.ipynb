**Recursion** in Tower of Hanoi

```python
def move_disks(n, source, auxiliary, destination):
    if n==1:
        print(f"Move disk 1 from {source} to {destination}")
        return
    move_disks(n-1, source, destination, auxiliary)
    print(f"Move disk {n} from {source} to {destination}")
    move_disks(n-1, auxiliary, source, destination)
```

```python
n = int(input("Enter numer of disks (n): "))
move_disks(n, "Source", "Auxiliary", "Destination")
```

**Iteration** in Tower of Hanoi, much less natural/intuitive/pythonic

```python
n = int(input("Enter number of disks (n): "))
src = "Source"
aux = "Auxiliary"
des = "Destination"

total_moves = 2**n - 1
if n % 2 == 0:
    des, aux = aux, des

rods = {src: list(range(n, 0, -1)), aux: [], des: []}

for move in range (1, total_moves + 1):
    if move % 3 == 1:
        from_rod, to_rod = (src, des) if (rods[src] and (not rods[des] or rods [src][-1] < rods [des][-1])) else (des, src)
    elif move % 3 == 2:
        from_rod, to_rod = (src, aux) if (rods[src] and (not rods[aux] or rods [src][-1] < rods [aux][-1])) else (aux, src)
    else: 
        from_rod, to_rod = (des, aux) if (rods[des] and (not rods[aux] or rods [des][-1] < rods [aux][-1])) else (aux, des)

    disk = rods[from_rod].pop()
    rods[to_rod].append(disk)
    print(f"Move disk {disk} from {from_rod} to {to_rod}")
```

**Recursion** in Fibonacci Numbers

```python
def fibonacci(n):
    """Returns the nth Fibnocci number"""
    if n == 0:
        return 0
    if n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
```

```python
n = int(input("Find the Fibonacci value of "))
fibonacci(n)
```

Increasing efficiency using **Memoization**

```python
def memofib(n, calls=None):
    """ Same as above, but uses calls to memoize previously computed values """
    if calls is None:
        calls = {}
    if n in calls:
        return calls[n]
    else:
        if n == 0:
            result = 0
        elif n == 1:
            result = 1
        else:
            result = memofib(n-1, calls) + memofib (n-2, calls)
        calls[n] = result
        return result
```
