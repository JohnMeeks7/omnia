```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
#from "Full Lab to master Seaborn" by Youcef Benaberrezak
```

```python
tips = sns.load_dataset("tips")
flights = sns.load_dataset("flights").pivot(index="month", columns="year", values="passengers")
penguins = sns.load_dataset("penguins")
```

```python
sns.histplot(data=tips["total_bill"], bins = 20, kde=True)
```

```python
sns.kdeplot(tips["total_bill"], fill = True)
```

```python
sns.rugplot(tips["total_bill"])
```

```python
sns.boxplot(data=tips, x="day", y="total_bill")
```

```python
sns.violinplot(data=tips, x="day", y="total_bill")
```

```python
sns.stripplot(data=tips, x="day", y="total_bill") #allows overlapping
```

```python
sns.swarmplot(data=tips, x="day", y="total_bill") #no overlapping, side-by-side
```

```python
sns.barplot(data=tips, x="day", y="total_bill")
```

```python
sns.countplot(data=tips,x="day")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue='day')
```

```python
sns.lineplot(x="size", y="tip", data=tips, errorbar="se") #size = table size, error bar possible arguments = sd, pi, ci, se standard deviation, percentile interval, confidence interval, standard error
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="day") #similar to scatterplot but allows for multiple plots, see below
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", kind="scatter", col="day", row="time")
```

```python
sns.heatmap(data=flights, annot=True, cmap="magma",fmt="d") #annot adds color bar on right, "d" means decimal integer, cmap = x is color, linewidths, linecolor, etc/
```

```python
sns.clustermap(data=flights, cmap="coolwarm")
```

```python
sns.pairplot(data=penguins, hue="species")
```

```python
sns.jointplot(data=tips, x="total_bill", y="tip", kind="scatter") #kind could be line, hex, kde, reg, hist
```

```python
sns.jointplot(data=tips, x="total_bill", y="tip", kind="hex") #kind could be line, hex, kde, reg, hist
```

```python
sns.jointplot(data=tips, x="total_bill", y="tip", kind="kde", fill="True") #kind could be line, hex, kde, reg, hist
```

```python
sns.regplot(data=tips, x="total_bill", y="tip")
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", hue="sex")
```
