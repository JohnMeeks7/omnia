```python
from darts import TimeSeries
import pandas as pd
```

```python
gold = pd.read_csv("Gold.csv")
#Daily gold fixing price (10:30 am London time) in London Bullion Market in US dollars per Troy ounce from January 3, 1995 to November 10, 2016.
```

```python
gold.head()
```

```python
from darts.utils.missing_values import fill_missing_values
series = TimeSeries.from_dataframe(gold, time_col="DATE", value_cols="VALUE", freq="B", fill_missing_dates=True)
series = fill_missing_values(series)
```

### Baseline Models

```python
#Naive Mean
from darts.models import NaiveMean
model = NaiveMean()
model.fit(series)
pred = model.predict(5)
print(pred)
```

```python
#Naive Seasonal
from darts.models import NaiveSeasonal
model = NaiveSeasonal(K=260) #how many of the last values to repeat, 260 for 260 business days
model.fit(series)
pred = model.predict(5)
print(pred)
```

```python
#Naive Drift
from darts.models import NaiveDrift
model = NaiveDrift()
model.fit(series)
pred = model.predict(5)
print(pred)
```

```python
#Naive Moving Average
from darts.models import NaiveMovingAverage
model = NaiveMovingAverage(input_chunk_length=10)
model.fit(series)
pred = model.predict(5)
print(pred)
```

### Basic Statistical Models: AR, ARMA, ARIMA

```python
#all just use ARIMA formula
from darts.models import ARIMA
```

```python
#AR
model = ARIMA(p=1, d=0, q=0)
model.fit(series)
pred = model.predict(5)
print(pred)
```

```python
#ARMA
model = ARIMA(p=1, d=0, q=1)
model.fit(series)
pred = model.predict(5)
print(pred)
```

```python
#ARIMA (p,d,q)
#ARMA
model = ARIMA(p=1, d=1, q=1)
model.fit(series)
pred = model.predict(5)
print(pred)
```

```python
from darts.models import ExponentialSmoothing
model = ExponentialSmoothing()
model.fit(series)
pred = model.predict(5)
print(pred)
```
