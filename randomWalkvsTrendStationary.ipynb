```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import pandas as pd
```

```python
n = 100
drift = .01
sigWt = 1
t = np.arange(1,101,1)
mu = drift * t

randomWalks = []
regressions = []
for i in range(4):
    wT = np.random.normal(0, sigWt, n)
    xT = (drift * t) + np.cumsum(wT)
    randomWalks.append(xT)
    model = sm.OLS(xT, t).fit()
    fittedLine = model.predict(t)
    regressions.append(fittedLine)
    plt.figure(figsize=(4,2))
    sns.lineplot(x=t, y=xT, label='Random Walk')
    sns.lineplot(x=t, y=fittedLine, label='Fitted Line')
    sns.lineplot(x=t, y=mu, linestyle='--', color='black')
    
```

```python
n = 100
drift = .01
sigWt = 1
t = np.arange(1,101,1)
mu = drift * t

trendStationary = []
regressions = []
for i in range(4):
    wT = np.random.normal(0, sigWt, n)
    xT = (drift * t) + (wT)
    trendStationary.append(xT)
    model = sm.OLS(xT, t).fit()
    fittedLine = model.predict(t)
    regressions.append(fittedLine)
    plt.figure(figsize=(4,2))
    sns.lineplot(x=t, y=xT, label='Stationary Trend')
    sns.lineplot(x=t, y=fittedLine, label='Fitted Line')
    sns.lineplot(x=t, y=mu, linestyle='--', color='black')
```

This exercise shows that in a random walk, the white noise is remembered overtime and large shocks cause large deviations that persist. More simply, variance scales with time. This causes vast discrepancies between the mean function and the fitted line. This is not true for stationary trends where white noise is unique to a single point and then forgotten, resulting in much smoother results and a regression line much closer to the true mean function.
