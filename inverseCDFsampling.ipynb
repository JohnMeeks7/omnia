```python
#Simulate random observations of any cdf (whose closed form inverse exists) using random pulls from a Uniform(0,1) distribution
```

```python
import numpy as np
```

```python
u = np.random.uniform(0,1, 10) #third value = # obs
```

```python
u #vector of n observations
```

```python
def inverseCDFsample(u, rate): 
    """ In this case, exponential dist. used with adjustable rate """
    return -np.log(1-u)/rate
```

```python
inverseCDFsample(u, 2)
```
