```python
#Nonparametric CI based on order statistics for 90th quantile
```

```python
n, p = 30, 0.9
l = int(stats.binom.ppf(0.025, n, p))
u = int(stats.binom.ppf(0.975, n, p))
lower = df.iloc[l-1, 0]
upper = df.iloc[u-1, 0]
print(f"95% CI using nonparametric order statistics for 90th (upper) quantile is {lower}, {upper}")
#Units are not specified in this practice data set
```

```python
#Construct a 95% prediction interval for the distance the next observed engine will last. If necessary, transform the data to Normality, then back-transform the interval endpoints.
```

```python
transformedDF = np.log(df[0])
mean =  np.mean(transformedDF)
std = np.std(transformedDF, ddof=1) #setting df to n-1 vs Numpy default of n df
tstat = stats.t.ppf(0.975, n-1)
se = std * np.sqrt(1 + (1/n))
lower = mean - tstat * se
upper = mean + tstat * se
#backtransform
lower = np.exp(lower)
upper = np.exp(upper)
print(f"95% prediction interval for the distance the next observed engine will last is {lower:.2f}, {upper:.2f}")
#Units are not specified in this practice data set
```

```python
#Construct an interval which will contain 95% of the population with 95% confidence
```

```python
#Tolerance interval
```

```python
#Using already transformed data
confidenceLevel = 0.95
proportionOfPopulation = 0.95
zscore = stats.norm.ppf((1+proportionOfPopulation)/2)
chisq = stats.chi2.ppf(1-confidenceLevel, n-1)
toleranceFactor = zscore * np.sqrt((n-1) * (1+1/n) /chisq) 
lower = mean - toleranceFactor * se
upper = mean + toleranceFactor * se
#backtransform
lower = np.exp(lower)
upper = np.exp(upper)
print(f"95% tolerance interval for 95% of the total popluation is {lower:.2f}, {upper:.2f}")
```

```python
#Construct a 95% CI for mean response time using i) bootstrap ii) parametric bootstrap
df2 = pd.read_csv("q2.csv", header=None)
```

```python
#original sample mean
xbar = df2.mean()
```

```python
#bootstrap
nsamples = 1000
nsize = len(df2)
bootstrapMeans = []
for sample in range(nsamples):
    resample = np.random.choice(df2[0], size=nsize, replace=True)
    bootstrapMeans.append(np.mean(resample))
bootstrapCI = np.percentile(bootstrapMeans, [2.5, 97.5])
print(f"95% bootstrap confidence interval for mean response time is {bootstrapCI}")
print(f"Compared to original sample mean of {xbar.iloc[0]}")
```

```python
#parametric bootstrap (expoenntial dist.)
#getting distribution parameter lambda
rate = 1/xbar
nsamples = 1000
parametricMeans = []
for sample in range(nsamples):
    resample = np.random.exponential(scale=1/rate, size=nsize)
    parametricMeans.append(np.mean(resample))
parametricCI = np.percentile(parametricMeans, [2.5, 97.5])
print(f"95% bootstrap confidence interval for mean response time using rate {rate.iloc[0]:.5f} is {parametricCI}")
```

```python
#confidence interval of i) a proportion < 30 months ii) median
df3 = pd.read_csv("q3.csv", header=None)
```

```python
#Proportion <30 months, using Wald interval
n = len(df3)
successes = np.sum(df3<30, axis=0)
phat = successes/n
se = np.sqrt(phat * (1-phat)/n)
zstat = 1.96 #n=60
l, u = phat - zstat*se, phat + zstat*se
print(f"95% Wald confidence interval for proportion of failures before 30 months is {l.iloc[0]:.4f},{u.iloc[0]:.4f} months")
```

```python
#Median using bootstrap
nsamples = 1000
nsize = n #len(df3)
bootstrapMedians = []

for sample in range(nsamples):
    resample = np.random.choice(df3[0], size=nsize, replace=True)
    bootstrapMedians.append(np.median(resample))
medianCI = (np.percentile(bootstrapMedians, [2.5, 97.5]))
print(f"95% bootstrap confidence interval for median of response times is {medianCI[0]:.2f}, {medianCI[1]:.2f} months")                     
```
