```python
#Quantile estimator (assumes sorted data)
def quantileSetosa(u):
    n = len(setosa)
    quantilePercent = (((n-1) * u) + 1)
    roundedQuantilePercent = round(quantilePercent)
    return setosa[roundedQuantilePercent - 1]

def quantileVirginica(u):
    n = len(virginica)
    quantilePercent = (((n-1) * u) + 1)
    roundedQuantilePercent = round(quantilePercent)
    return virginica[roundedQuantilePercent - 1]
```

```python
print(quantileSetosa(.1))
print(quantileSetosa(.4))
print(quantileSetosa(.8))
```

```python
print(quantileVirginica(.1))
print(quantileVirginica(.4))
print(quantileVirginica(.8))
```
